```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\nstow\Desktop\Python\suicideData.csv")
```

## Suicides by Year and Age Ploted 4 Different Ways

```python
dfAge5To14 = df.loc[(df['age']=='5-14 years')]
dfAge5To14 = dfAge5To14.groupby(['year']).sum()
dfAge5To14 = dfAge5To14.reset_index()

dfAge15To24 = df.loc[(df['age']=='15-24 years')]
dfAge15To24 = dfAge15To24.groupby(['year']).sum()
dfAge15To24 = dfAge15To24.reset_index()

dfAge25To34 = df.loc[(df['age']=='25-34 years')]
dfAge25To34 = dfAge25To34.groupby(['year']).sum()
dfAge25To34 = dfAge25To34.reset_index()

dfAge35To54 = df.loc[(df['age']=='35-54 years')]
dfAge35To54 = dfAge35To54.groupby(['year']).sum()
dfAge35To54 = dfAge35To54.reset_index()

dfAge55To74 = df.loc[(df['age']=='55-74 years')]
dfAge55To74 = dfAge55To74.groupby(['year']).sum()
dfAge55To74 = dfAge55To74.reset_index()

dfAge75 = df.loc[(df['age']=='75+ years')]
dfAge75 = dfAge75.groupby(['year']).sum()
dfAge75 = dfAge75.reset_index()
print(df)
```

```python
plt.figure(figsize = (15,6))

plt.plot(dfAge5To14['year'], dfAge5To14['suicides_no'], color = 'red')
plt.plot(dfAge15To24['year'], dfAge15To24['suicides_no'], color = 'green')
plt.plot(dfAge25To34['year'], dfAge25To34['suicides_no'], color = 'orange')
plt.plot(dfAge35To54['year'], dfAge35To54['suicides_no'], color = 'blue')
plt.plot(dfAge55To74['year'], dfAge55To74['suicides_no'], color = 'purple')
plt.plot(dfAge75['year'], dfAge75['suicides_no'], color = 'magenta')
```

```python
plt.figure(figsize = (15,6))
plt.subplot(2,3,1)
plt.title('Suicides for 5-14')
plt.ylim([0, 110000])
plt.plot(dfAge5To14['year'], dfAge5To14['suicides_no'], color = 'red')
plt.subplot(2,3,2)
plt.title('Suicides for 15-24')
plt.ylim([0, 110000])
plt.plot(dfAge15To24['year'], dfAge15To24['suicides_no'], color = 'green')
plt.subplot(2,3,3)
plt.title('Suicides for 25-34')
plt.ylim([0, 110000])
plt.plot(dfAge25To34['year'], dfAge25To34['suicides_no'], color = 'orange')
plt.subplot(2,3,4)
plt.title('Suicides for 35-54')
plt.ylim([0, 110000])
plt.plot(dfAge35To54['year'], dfAge35To54['suicides_no'], color = 'blue')
plt.subplot(2,3,5)
plt.title('Suicides for 55-74')
plt.ylim([0, 110000])
plt.plot(dfAge55To74['year'], dfAge55To74['suicides_no'], color = 'purple')
plt.subplot(2,3,6)
plt.title('Suicides for 75+')
plt.ylim([0, 110000])
plt.plot(dfAge75['year'], dfAge75['suicides_no'], color = 'magenta')
```

```python
plt.figure(figsize = (15,6))
ax = sns.scatterplot(x="year", y="suicides_no", hue='age', data=df)
plt.legend(loc='upper right')
```

```python
g = sns.FacetGrid(df, col="age", hue="age",height=4, aspect=1, col_wrap=3)
g.map(plt.scatter, "year", "suicides_no", alpha=.7)
g.add_legend()
```

## Two Continuous Variables Plotted 3 Different Ways

```python
plt.figure(figsize = (5,6))
plt.scatter(df['suicides_no'], df['population'], color ='green')
```

```python
g = sns.lmplot(x='suicides_no',
                y='population', 
                data=df,
                fit_reg=True, 
                scatter_kws={'alpha':0.4})
plt.figure(figsize = (5,6))
g.set_xlabels("Suicide Numbers")
g.set_xlabels("Total Population")
plt.show()
```

```python
# Setting the overall aesthetic.
sns.set(style="white")

g = sns.lmplot(x='suicides_no',
                y='population',
               hue='age',
               data=df,
               fit_reg=True,
               ci=False,
               scatter_kws={'alpha':0.4})
plt.figure(figsize = (5,6))
g.set_ylabels("Population")
g.set_xlabels("Suicide Numbers")
plt.subplots_adjust(top=0.9)
plt.show()
```
