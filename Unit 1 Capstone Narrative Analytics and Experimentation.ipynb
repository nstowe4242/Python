```python
df = df.drop(df.index[65])
```

```python
print(df.describe())
```

```python
plt.figure(figsize = (15,3))
plt.xticks(fontsize = 6, rotation = 90)
plt.title('Total Fentanyl Prescriptions Dispensed')
plt.plot(df['month_dispensed'], df['all_drug_total'])
```

We can see an overall drop in the number of Fentanyl prescriptions that were dispensed over the five-year period. To ensure that the trend is across all the different Fentanyl containing drugs we will take a look at all of the drugs individually.

```python
plt.figure(figsize = (15,20))

plt.subplot(4,2,1)
plt.xticks(fontsize = 4, rotation = 90)
plt.title('Abstral Prescriptions Dispensed')
plt.plot(df['month_dispensed'], df['abstral'], color = 'orange')

plt.subplot(4,2,2)
plt.xticks(fontsize = 4, rotation = 90)
plt.title('Actiq Prescriptions Dispensed')
plt.plot(df['month_dispensed'], df['actiq'], color = 'red')

plt.subplot(4,2,3)
plt.xticks(fontsize = 4, rotation = 90)
plt.title('Duragesic Prescriptions Dispensed')
plt.plot(df['month_dispensed'], df['duragesic'], color = 'purple')

plt.subplot(4,2,4)
plt.xticks(fontsize = 4, rotation = 90)
plt.title('Fentanyl Prescriptions Dispensed')
plt.plot(df['month_dispensed'], df['fentanyl'], color = 'blue')

plt.subplot(4,2,5)
plt.xticks(fontsize = 4, rotation = 90)
plt.title('Fentanyl Prescriptions Dispensed')
plt.plot(df['month_dispensed'], df['fentanyl_citrate'], color = 'green')

plt.subplot(4,2,6)
plt.xticks(fontsize = 4, rotation = 90)
plt.title('Abstral Prescriptions Dispensed')
plt.plot(df['month_dispensed'], df['fentora'], color = 'yellow')

plt.subplot(4,2,7)
plt.xticks(fontsize = 4, rotation = 90)
plt.title('Lazanda Prescriptions Dispensed')
plt.plot(df['month_dispensed'], df['lazanda'], color ='brown')

plt.subplot(4,2,8)
plt.xticks(fontsize = 4, rotation = 90)
plt.title('Aubsys Prescriptions Dispensed')
plt.plot(df['month_dispensed'], df['subsys'], color = 'maroon')
```

From these plots we can see that not all the different drugs are seeing an overall downward trend. 
Abstral, a sublingual dissolving tablet, saw a sharp decline from February 2014 to July 2014. However, there was a resurgence in dispensed prescriptions between February 2016 to January 2017.
Additionally, Lazanda, which is a nasal spray, has seen an overall increase in dispensed prescriptions.

# Scope of Experiment

With this data we would wonder why, if there was an overall decrease is Fentanyl prescriptions that were dispensed, do we often still hear about an increase in opioid deaths? This is the cornerstone of our experiment. The question we will set out to answer is for months that there was a decrease in overall Fentanyl prescriptions that were dispensed, was there also an increase in the number of other opioid prescriptions that were dispensed.

To answer this question, we will need to gather the monthly number of all other opioid prescriptions that were dispensed. 

It will also be useful to investigate the number of heroin deaths. This is important because we can't track the amount that is dispensed since heroin is an illegal substance. We can track the number of deaths due to heroin through the CDC. Along with this we will look at deaths from prescription opioids. With this additional information, we can see the impact of the number of prescription opioids that are dispensed and the possible correlation in the use of and deaths from heroin (as an alternative to prescription opioids).

# Hypothesis

The trend in prescription opiods that are dispensed will follow a downward trend that is similar to the downward trend in Fentanyl dispensing.
